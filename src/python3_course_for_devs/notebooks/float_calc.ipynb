# 浮点数计算


```python
# 开门见山，还是这道题，为什么？

print(0.1 + 0.2 != 0.3)  # True
```

首先，我们知道，计算机只认识二进制，我们平时写的十进制在计算机里都要转为二进制才能进行计算，
但是，由于十进制对于我们来说更加直观，且与我们通常使用的数值概念更为一致，因此在数值比较的时候，还是会用十进制
所以会出现一种情况，就是一个看起来精确的十进制小数，实际上是一个二进制无限循环小数，也就是不精确值。

好了，我们下面开始分析这个问题

首先看一下十进制如何转二进制，这里分两种各种情况：

1. 整数十进制转二进制

整数十进制转二进制采用的是除 2 取余法（短除法）

比如整数 13 的二进制转换，步骤如下

|      | 商  | 余  |
| ---- | --- | --- |
| 13/2 | 6   | 1   |
| 6/2  | 3   | 0   |
| 3/2  | 1   | 1   |
| 1/2  | 0   | 1   |

结果为 1101，看起来还算直观吧

2. 小数十进制转二进制

小数十进制转二进制采用的是乘 2 取整法

比如小数 0.1 的二进制转换，步骤如下

|          | 整数部分 | 小数部分 |
| -------- | -------- | -------- |
| 0.1 \* 2 | 0        | 0.2      |
| 0.2 \* 2 | 0        | 0.4      |
| 0.4 \* 2 | 0        | 0.8      |
| 0.8 \* 2 | 1        | 0.6      |
| 0.6 \* 2 | 1        | 0.2      |
| 0.2 \* 2 | 0        | 0.4      |
| 0.4 \* 2 | 0        | 0.8      |
| 0.8 \* 2 | 1        | 0.6      |
| 0.6 \* 2 | 1        | 0.2      |

结果为 0.00011001100... 无限循环

因此，你知道了，0.1 由十进制转为二进制的时候结果是个无限循环小数，并不是一个精确值

同样的，0.2 转为二进制为 0.00110011001...

结果呢，0.1 + 0.2 同样是一个无限循环小数

我们已经知道了，数字运算的时候当它被转为十进制的时候精度会丢失，十进制近似值为 0.30000000000000004，

显然这个值和 0.3 不相等

但是你可能还会有疑惑，既然 0.1 + 0.2 是个无限循环小数，0.3 的二进制也是个无限循环小数，而且看起来，它们是一样的无限循环小数，
那么为什么不用二进制直接进行对比，而要转为十进制呢，那是因为在计算机中，就算是二进制也是用有限的位数进行存储的，一样存在舍入误差


```python
print(0.1 + 0.2)
print(0.3)
```
